```python
plt.plot(x,y,"red") #dosyalarda kullanılırsa plt.show() kullanılmak zorundadır.
plt.show()
```

```python
plt.subplot(2,2,1) #2 ye 2 lik düzlemde 1. grafik
plt.plot(x,y,"blue")
plt.subplot(2,2,2)#2 ye 2 lik düzlemde 2. grafik
plt.plot(y,x,"yellow")
plt.subplot(2,2,3)#2 ye 2 lik düzlemde 3. grafik
plt.plot(x**x,y,"red")
plt.subplot(2,2,4)#2 ye 2 lik düzlemde 4. grafik
plt.plot(x,y**y,"black")
```

```python
#FİGÜR OLUŞTURMA
x = np.arange(1,6)
y = np.arange(2,11,2)

fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.2,0.3])

axes1.plot(y,x)
axes1.set_xlabel("Outer X")
axes1.set_ylabel("Outer Y")
axes1.set_title("Outer Graph")

axes2.plot(x,y)
axes2.set_xlabel("Inner X")
axes2.set_ylabel("Inner Y")
axes2.set_title("Inner Graph")
plt.show()
```

```python
#GRAFİKLERİ BİRAZ DAHA ÖZELLEŞTİRME
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,color="red",linewidth=3,linestyle ="--",marker ="o",markersize=3
          ,markerfacecolor ="black",markeredgecolor="purple",markeredgewidth =15) 
#ÇİZGİ KALINLIĞI İÇİN "linewidth" veya"lw" 
#linestyle ise çizginin şeklini ayarlar "ls" şeklinde de yazılabilir.
#marker değerlerin işaretlenmesini sağlar işaretin büyüklüğü içinse "markersize" kullanılır. 
#Rengi için markerfacecolor metodu kullanılır.
#marker ların belirginleştirilmesi arka fonla olabilir bunun için "markeredgecolor", boyunu için ise "markeredgewidth" kullanılır 
axes.set_xlim(0,20)
axes.set_ylim(0,20)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(51)
y = np.arange(0,101,2)
z = x ** 3
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,"red")
axes.set_xlabel("X Values")
axes.set_ylabel("Y Values")
axes.set_title("X vs Y Graph")
plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
axes1=fig.add_axes([0,0.1,0.8,0.8])
axes2=fig.add_axes([0.1,0.5,0.4,0.3])
axes1.plot(x,z,"blue",lw=4,ls="--",marker = "o",markersize=5,markerfacecolor="black"
           ,markeredgecolor="black",markeredgewidth=0.1)
axes1.set_xlabel("Inner X Values")
axes1.set_ylabel("Inner Y Value")
axes1.set_title("Inner Graph")
axes2.plot(x,y,"red",lw=4,ls=":",marker="o",markersize=5,markerfacecolor="purple"
           ,markeredgecolor="purple",markeredgewidth=0.5)
axes2.set_xlabel("Order X Values")
axes2.set_ylabel("Order Y Values")
axes2.set_title("Order Graph")
axes2.plot(x,y,"red")
plt.show()
```

```python
fig,axes = plt.subplots(nrows =1 ,ncols = 2,figsize=(10,6))
axes[0].plot(x,y,"red",linewidth = 3, linestyle = ":",marker="o",markerfacecolor = "black")
axes[0].set_xlabel("X Values")
axes[0].set_ylabel("Y Values")
axes[0].set_title("X vs Y Graph")
axes[1].plot(x,z,"purple",lw=3,ls="--")
axes[1].set_xlabel("X Values")
axes[1].set_ylabel("Z Values")
axes[1].set_title("X vs Z Graph")
plt.show()
```
